```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle


```

```python
with open("investors_df.pkl", 'rb') as picklefile: 
    df_investors = pickle.load(picklefile)
```

```python
df_investors.head()
```

```python
df_investors.describe()
```

```python
# From describe we get the quantiles of 1.63*10^7, 2.6*10^6, 5.0*10^5 and 0.

def get_ranking(x):
    if x > 1.63*10**7:
        return 1
    elif x > 2.6*10**6:
        return 2
    elif x > 5.0*10**5:
        return 3
    else: 
        return 4
```

```python
df_investors['ranking'] = df_investors['total_investment'].map(get_ranking)
```

```python
with open("nazim_initial_df.pkl", 'rb') as picklefile: 
    df_comp = pickle.load(picklefile)
```

```python
df_comp
```

```python
def average_ranking(x_list):
    total_ranking = 0 
    
    try:
        if len(x_list)>0:  
            for i in x_list:
                total_ranking += df_investors.loc[i,'ranking']
        else:
            return 4

        return total_ranking/len(x_list)
    
    except TypeError:
        return 4
        
    
```

```python
df_comp['avg_investor_rank'] = df_comp['investors'].map(average_ranking)
```

```python
df_comp
```

```python
with open("cities_df.pkl", 'rb') as picklefile: 
    df_cities = pickle.load(picklefile)
```

```python
def get_city_investment(x):
    try:
        return df_cities.loc[x,'total_invested']
    except KeyError:
        return 0
```

```python
df_comp['city_investment'] = df_comp['city'].map(get_city_investment)
```
