```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load excel

xls = pd.ExcelFile('crunchbase_monthly_export_d43b4klo2ade53.xlsx')
```

```python
# Create a dataframe of all acquisitions
df_acq = pd.read_excel(xls, 'Acquisitions')

# List of all companies that have been acquired
acquired_list = list(df_acq['company_name'])

acquired_list
```

```python
# Load in all companies and set company name as index

df_comp = pd.read_excel(xls, 'Companies')
df_comp = df_comp.set_index('name')
```

```python
# Function to check whether a company has been acquired or not.

def acquired_check(i):
    if i in acquired_list:
        return 1
    else: 
        return 0
```

```python
# Create new column for acquired or not

df_comp['acquired'] = df_comp.index.map(acquired_check)

df_comp.head()
```

```python
# Load in investment data

df_invest = pd.read_excel(xls, 'Investments')

df_invest = df_invest.set_index('company_name')
df_invest = df_invest.rename({'fILIPPO Murroni':'investor'},axis='columns')
df_invest = df_invest.drop(index='test company')

df_invest.head()
```

```python
# Create list of all investors

investor_list = df_invest.investor.unique()
investor_list
```

```python
# Create dataframe for investors

investor_dict_list = []

for i in investor_list:
    temp_df = df_invest[df_invest['investor']== i]
    investor_dict_list.append({'investor':i,'invested_companies':list(temp_df.index),'no_investments':len(list(temp_df.index)),\
                        'total_investment':temp_df['raised_amount_usd'].sum()})
    
df_investors = pd.DataFrame(investor_dict_list)
df_investors = df_investors.set_index('investor')
df_investors
```

```python
# Create dataframe for all cities

city_list = df_invest.company_city.unique()
city_dict_list = []

for i in city_list:
    temp_df = df_invest[df_invest['company_city']==i]
    city_dict_list.append({'city':i, 'total_invested_companies': len(list(temp_df.index)), \
                           'total_invested': temp_df['raised_amount_usd'].sum()})

df_cities = pd.DataFrame(city_dict_list)
df_cities = df_cities.set_index('city')

df_cities
```
