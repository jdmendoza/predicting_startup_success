```python
data_df = df1[['name','first_funding_at','last_funding_at']]
```

```python
data_df.reset_index();
```

```python
from datetime import datetime
from datetime import date

def find_date(df_elem):
    date1 = df_elem.split(' ')[0].split('-')
    date1 = date(int(date1[0]),int(date1[1]),int(date1[2]))
    return date1
```

```python
data_df.head(5)
```

```python
timedelta_tft = (data_df['last_funding_at'] - data_df['first_funding_at']);
data_df['funding_total_time'] = [int(x.days) for x in list(timedelta_tft)]
data_df.reset_index();
```

```python
data_df.head(10)
```

```python
total_funding_time_dict = dict(zip(list(data_df.name),list(data_df.funding_total_time)));
```

```python
import pickle

with open('sf_total_funding_time_dict.pickle', 'wb') as handle:
    pickle.dump(total_funding_time_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
